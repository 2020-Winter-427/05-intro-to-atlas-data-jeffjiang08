In this part of the assignment, we going to create histograms from the number of leptons.


First, we have to open the root file with data that we want to analyze using uproot. 

```python
f = uproot.open("https://atlas-opendata.web.cern.ch/atlas-opendata/samples/2020/1largeRjet1lep/MC/mc_361106.Zee.1largeRjet1lep.root")
```

Next, we take the data of Number of leptons out of the root file.

```python
x = f["mini/lep_n"].array()
```

Graph the number of lepton as a histogram.

```python
fig, ax = plt.subplots(1, 1)
ax.hist(x,3)
plt.title("Example plot: Number of leptons")
plt.xlabel('Number of leptons')
plt.ylabel('Events')
```

Graph the same histogram but normalized. 

```python
fig, ax = plt.subplots(1, 1)
ax.hist(x,3,density = True)
plt.title("Example plot: Number of leptons")
plt.xlabel('Number of leptons')
plt.ylabel('Events')
```
